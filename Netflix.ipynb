```python
col_to_replace=['director' , 'cast' , 'country']
df[col_to_replace]=df[col_to_replace].replace(np.NaN , 'unknown')
df.head()
```

```python
col_to_drop=['date_added' , 'rating' , 'duration']
df.dropna(subset=col_to_drop , inplace=True)
df.sample(5)
```

```python
df['date_added'] = df['date_added'].astype('datetime64[ms]')
df.dtypes
```

```python
df.head()
```

```python
df['release_year'].unique()
```

```python
df['year_added']=df['date_added'].dt.year
print(df['year_added'])
```

```python
df['month_added']= df['date_added'].dt.month_name()
df['month_added']
```

```python
pd.crosstab(df.type , df.year_added)
```

```python
pd.crosstab(df.type , df.year_added).plot(kind='bar')
plt.tight_layout()
plt.xlabel("Content Type", fontweight ='bold', fontsize=12)
plt.legend(title="Year Added", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Netlix Content by Type and Year Added', fontweight='bold', fontsize =12)
plt.grid(True,which= 'both',linestyle='--', linewidth=0.2)
plt.xticks(rotation=2)
plt.show()
```

```python
df.groupby(['year_added' , 'type']).size().reset_index(name='count').sort_values(by='count',ascending= False)
```

```python
year_type_count = df.groupby(['year_added', 'type']).size().reset_index(name='count')
sns.lineplot(data=year_type_count, x='year_added', y='count', hue='type',palette='Set2')
plt.title("Content Added Over the Years by Type")
plt.xlabel('Year Added')
plt.grid(True)
plt.show()
```

```python
plt.tight_layout()
sns.histplot(data=df, x= 'year_added', bins=20, kde=True, color="darkcyan" )
plt.title("Distribution of Netflix Content by Year Added")
plt.xlabel("Year Added")
plt.xticks(rotation=45)
plt.grid(True, which='both', axis='y', linestyle='-.', linewidth=0.7)
plt.show()
```

### Countries which have the most content available on Netflix

```python
top10_countries = df['country'].replace('unknown',pd.NA).dropna().value_counts().head(10)
top10_countries
```

```python
colors = sns.color_palette("Blues", len(top10_countries))[::-1]
top10_countries.plot(kind='barh', color = colors)
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Country")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
year_country = df.groupby(['year_added','country']).size().reset_index(name='count').sort_values("count",ascending=False)
year_country
```

```python
country_year_trends = df.groupby(['year_added', 'country']).size().reset_index(name='count')
top_countries = country_year_trends[country_year_trends['country'].isin(['United States', 'India', 'United Kingdom', 'Japan', 'South Korea'])]
sns.lineplot(data=top_countries, x='year_added', y='count', hue='country',palette='Set2')
plt.title("Content Added Over the Years by Type")
plt.xlabel('Year Added')
plt.grid(True)
plt.show()
```

```python
top10_list_in = df['listed_in'].str.split(',').explode().value_counts().head(10)
colors = sns.color_palette("Purples", len(top10_list_in))[::-1]
plt.figure(figsize=(10,5))
top10_list_in.plot(kind='bar',color= colors)
plt.title("Top 10 Most Popular Genres on Netflix", fontsize=16)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

### Top Directors

```python
top10_director = df['director'].replace('unknown',pd.NA).dropna().value_counts().head(10)
top10_director
```

```python
plt.figure(figsize=(10,6))
colors= sns.color_palette('Reds',len(top10_director))[::-1]
top10_director.plot(kind='bar',color = colors)
plt.title("Top 10 Directors on Netflix", fontweight='bold',fontsize=16)
plt.xlabel("Director", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout()
plt.show()
```

### Finding the average duration of Movies

```python
movie_data = df[df['type'] == "Movie"]
avg_movie_duration = movie_data['duration'].str.replace(' min', '').astype(int).mean()
print(f"Average Movie Duration : {avg_movie_duration:.2f} minutes")
```

```python
movie_duration = df[df['type']=='Movie']['duration'].str.replace(' min','').astype(int)
sns.histplot(movie_duration, bins=20, kde=True, color='skyblue')
plt.title("Distribution of Movie Duration on Netflix")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")
plt.show()
```
